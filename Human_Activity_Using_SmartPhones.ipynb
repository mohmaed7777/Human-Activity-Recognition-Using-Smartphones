```python
# We do need to cinvert the Acitivty column from string base into integers by using LabelEncoder() function : 
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['Activity'] = le.fit_transform(data.Activity)
```

```python
data['Activity'].sample(5)
```

```python
# Calculate the correlation between the independent Variables : 
features_cols = data.columns[:-1]
corr_values = data[features_cols].corr()
corr_values
```

```python
# Simplify by emptying all the data below the diagnoal : 
tril_index = np.tril_indices_from(corr_values)
tril_index
```

```python
# Make the unused values as NaNs : 
corr_array = np.array(corr_values)
corr_array[np.tril_indices_from(corr_values)] = np.nan
pd.DataFrame(corr_array)
```
